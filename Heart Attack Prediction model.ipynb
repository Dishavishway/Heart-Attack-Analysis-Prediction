### 4.1 Uni-variate Analysis <a id = 18></a>

#### 4.1.1 Numerical Variables(Analysis with Distplot) <a id=19></a>

```python
numeric_var
```

```python
numeric_axis_name = ["Age of the Patient", "Resting Blood Pressure", "Cholesterol", "Maximum Heart Rate Achieved", "ST Depression"]
```

```python
numeric_var_with_axis = list(zip(numeric_var, numeric_axis_name))
```

```python
numeric_var_with_axis
```

```python
title_font = {"family": "arial", "color": "darkred", "weight": "bold", "size": 15}
axis_font = {"family": "arial", "color":"darkblue", "weight": "bold", "size": 13}

for i, z in numeric_var_with_axis:
    plt.figure(figsize = (8,6), dpi = 80)
    sns.distplot(df[i], hist_kws = dict(linewidth = 1, edgecolor = "k"), bins=20)
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict = axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

#### 4.1.2 Categorical Variables(Analysis with Pie Chart) <a id=21></a>

```python
categoric_var
```

```python
categoric_axis_name = ["Gender", "Chest Pain Type", "Fasting Blood Suger", "Resting Electrocardiographic Results", "Exercise Induced Angina", "The Slope of ST Segment", "Number of Major Vessels", "Thal", "Target"]
```

```python
categoric_var_with_axis = list(zip(categoric_var, categoric_axis_name))
```

```python
categoric_var_with_axis 
```

```python
df["cp"].value_counts()
```

```python
print(list(df["cp"].value_counts()))
print(list(df["cp"].value_counts().index))
```

```python
title_font = {"family": "arial", "color": "darkred", "weight": "bold", "size": 15}
axis_font = {"family": "arial", "color":"darkblue", "weight": "bold", "size": 13}

for i, z in categoric_var_with_axis:
    fig, ax = plt.subplots(figsize = (5, 4))
    
    observation_values = list(df[i].value_counts().index)
    total_observation_values = list(df[i].value_counts())
    
    ax.pie(total_observation_values, labels = observation_values, autopct = '%1.1f%%', startangle = 110, labeldistance = 1.1)
    ax.axis("equal")
    
    plt.title((i + "(" + z + ")"), fontdict = title_font)
    plt.legend()
    plt.show()
```
